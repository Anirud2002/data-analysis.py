```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 11)
```

```python
x
```

```python
plt.figure(figsize=(12, 6))

plt.title('First plot')

plt.plot(x, x)

plt.plot(x, x ** 2)

plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('Second plot')

plt.subplot(1, 2, 1) # rows, columns, panel selected
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0 , 100])
plt.legend(['X^2', 'Vertical Line'])

plt.xlabel('X')
plt.ylabel('X squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X squared')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes
```

```python
axes.plot(
    x, x ** 2, color='red', linewidth=3, marker='o', 
    markersize=5, label='X^2'
)

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X^2')


axes.set_title('Third plot using oop')
axes.legend()


fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted')

axes.set_title('Fourth plot')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '--og', label='solid green')
axes.plot(x, x + 1, '--c', label='dashed cyan')
axes.plot(x, x + 2, '-.b', label='dashdot blue')
axes.plot(x, x + 3, ':r', label='dotted red')

axes.legend()

axes.set_title('fifth plot')
```

### Scatter plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii
```

```python
fig, axes = plt.subplots(figsize=(14, 6))


axes.scatter(x, y, c=colors, s=area, alpha=0.6, cmap='Spectral')


```
